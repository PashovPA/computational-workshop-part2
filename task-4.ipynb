Найти максимальное по модулю собственное значение
матрицы степенным методом и методом скалярных произведений

```python
import numpy as np
import math
import copy

A = np.array([[-1.14896, -0.53716, 0.78959],
              [-0.53716, 0.88917, 0.19536],
              [0.78959, 0.19536, -1.28186]])

e = 10 ** (-3)
v = [1, 0, 0]
y = [v]
lambs = [v]

w, v = np.linalg.eig(A)
l_max = 0
for l in w:
  if abs(l)>abs(l_max):
    l_max = l

i = 1
l_1, l_2 = 1, 0

while np.abs(l_1 - l_2) > e:
  y.append(np.dot(A, y[i - 1]))
  l_2 = copy.deepcopy(l_1)
  l_1 = y[i][0] / y[i - 1][0]
  i+=1

w_max = l_1
y[i-1] = y[i-1]/np.linalg.norm(y[i-1])

print("Максимальное по модулю собственное число степенным методом:", w_max)
print("Соответствующие приближение собственного вектора:", y[i-1])
print("\nМаксимальное по модулю собственное число встроенным методом:", l_max)
print("Абсолютная фактическая погрешность:", abs(w_max - l_max))
print("Число итераций:", i)
```
